```python
# Printing initial shape and size of the dataset
print("Initial Shape and Size of the Dataset:")
print("Rows:", car_df.shape[0])
print("Columns:", car_df.shape[1])
```

```python
# Selecting numeric columns for mean imputation
car_df_numeric_missing = car_df.select_dtypes(include=['float64', 'int64'])
# Using mean imputation for numeric variables
car_df[car_df_numeric_missing.columns] = car_df_numeric_missing.fillna(car_df_numeric_missing.mean())
# Printing final shape and size of the cleaned dataset
print("\nFinal Shape and Size of the Cleaned Dataset:")
print("Rows:", car_df[car_df_numeric_missing.columns].shape[0])
print("Columns:", car_df[car_df_numeric_missing.columns].shape[1])
```

```python
# Drop duplicates if any
print("Dropping duplicates")
car_df = car_df.drop_duplicates()
# Counting the number of duplicate rows
duplicates_count = car_df.duplicated().sum()
# Printing DataFrame after handling missing values and duplicates
print("\n DataFrame after handling missing values and duplicates:")
print(car_df.info())
print("\nNumber of duplicate rows removed:", duplicates_count)
```

```python
# Visualization of the variable Year using Histogram
# Set the figure size
plt.figure(figsize=(10, 6))
# Create a histogram with kernel density estimation (kde)
sns.histplot(car_df["Year"],kde=True)
# Set the title and axis labels
plt.title(f'Distribution of Year')
plt.xlabel('Year')
plt.ylabel('Frequency')
# Display the plot
plt.show()
```

### **Histogram (Distribution of Year):**

**Observation**:

The histogram illustrates the distribution of vehicle manufacturing years in the dataset. The histogram shows the distribution of vehicle manufacturing years in the dataset. Most vehicles were manufactured around 2017(between 2015 and 2020), with fewer vehicles in earlier years(1990 to 2005) and later years. The median year is 2017, indicating that half of the vehicles were manufactured before or after that year.


Variables Plotted: '**year**'


**Description**:

The x-axis represents the manufacturing years of vehicles, and the y-axis represents the frequency of different manufacturing years.

**Relationship/Information**:
    
The histogram provides insights into the distribution of vehicle
manufacturing years. It allows us to observe whether the dataset is skewed towards newer or older vehicles, providing context on the age distribution of the vehicles.



```python
# Measuring the Skewness and Modality of the 'Year' variable
# calculates the skewness of "Year"
skewness = car_df["Year"].skew()
# Determine the modality based on the skewness value
modality = "Positively Skewed" if skewness > 0 else "Negatively Skewed" if skewness < 0 else "Symmetric"
# Displays the Skewness and Modality
print(f"Skewness: {skewness}")
print(f"Modality: {modality}")
```

```python
# Analyze Outliers using Boxplot
# Set the figure size
plt.figure(figsize=(10, 6))
# Create a boxplot to visualize the distribution of 'Year'
sns.boxplot(x="Year", data=car_df)
plt.title('Boxplot for Year')
plt.xlabel(" Year ")
# Display the plot
plt.show()
```

```python
# Generating Correlation Heatmap for Numeric Variables
# Selecting numeric columns and creating a correlation matrix
numeric_columns = car_df.select_dtypes(include=['number']).columns
numeric_df = car_df[numeric_columns]
correlation_matrix = numeric_df.corr()
# Use Seaborn's heatmap for correlation visualization
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap for Numeric Variables")
plt.show()
```

A heatmap is a graphical representation of data in two dimensions, using colors to depict values. In the context of a Correlation Heatmap, it visually displays the degree of correlation, or dependence, between variables through color intensity, with different shades indicating the strength of correlation.

The given heatmap shows the correlation between all numerical variables in the dataframe. Red colors indicate positive correlations, while blue colors indicate negative correlations. The intensity of the color reflects the strength of the correlation.

This heatmap plots the correlations between all numeric variables present in the numeric_df DataFrame. By definition, a correlation measures the linear relationship between two variables, indicating whether they tend to increase or decrease together (positive correlation) or move in opposite directions (negative correlation).
